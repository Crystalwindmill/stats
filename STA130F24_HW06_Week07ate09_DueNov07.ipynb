````python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf  # for statistical modeling, statsmodels package and is used for estimating and testing statistical models. formula is a interface for relationships betweeen variables
import plotly.express as px  # for plotting

# Parameters for the simple linear regression
beta_0 = 2  # Intercept
beta_1 = 1.5  # Slope
sigma = 1  # Standard deviation of the error term

# Generate predictor values (x)
np.random.seed(0)  # For reproducibility
x = np.linspace(0, 10, 100)  # Creates 100 evenly spaced values from 0 to 10

# Generate error term (epsilon) as random noise from a normal distribution
# Ordinary Least Squares (OLS) regression model where Y is the dependent variable and x is the independent variable.
epsilon = np.random.normal(loc=0, scale=sigma, size=len(x))  # Generate noise with mean=0 and std=1

# Calculate outcome values (y) based on the linear regression model
y = beta_0 + beta_1 * x + epsilon  # y = beta_0 + beta_1 * x + epsilon

# Create a DataFrame for better visualization and handling
df = pd.DataFrame({'x': x, 'y': y})

# Create a linear regression model and fit it
# Ordinary Least Squares (OLS) regression model where Y is the dependent variable and x is the independent variable.
model_data_specification = smf.ols("y ~ x", data=df)  # Specify model: y = beta_0 + beta_1 * x
fitted_model = model_data_specification.fit()  # Fit the model to the data(df) 

# Model Summary
print(fitted_model.summary())  # Provides summary statistics of the model

# Accessing specific parts of the model output:
print(fitted_model.summary().tables[1])  # Provides the coefficient table
print(fitted_model.params)  # The parameters (intercept and slope) of the model, returns the estimated coefficients of the regression model as a pandas Series, including the intercept and slope(s).
print(fitted_model.params.values)  # The parameter values as a numpy array
print(fitted_model.rsquared)  # R-squared value of the model,indicating the proportion of variance in the dependent variable that can be explained by the independent variable(s).
#  An R-squared value closer to 1 indicates a better fit.

# Now, we create a plot using plotly express
df['Data'] = 'Data'  # Hack to add data to legend
fig = px.scatter(df, x='x', y='y', color='Data', 
                 trendline='ols', title='Y vs. x')  # Adding the trendline (OLS) to the plot

# This is essentially what `trendline='ols'` does:
fig.add_scatter(x=df['x'], y=fitted_model.fittedvalues,
                line=dict(color='blue'), name="OLS trendline")  # Add the fitted regression line

fig.show(renderer="png")  # Show the interactive plot
````

Here's a summary of our conversation on Simple Linear Regression and the interpretation of \( R^2 \) and correlation:

1. **Predictor and Outcome Variables**:
   - In a regression model, \( x \) is the predictor (independent variable), not the predicted value. The observed outcome (dependent variable) is \( Y \), and the model’s predicted values are denoted by \( \hat{Y} \).

2. **Calculation of \( R^2 \)**:
   - \( R^2 \) (the coefficient of determination) is calculated as:
     \[
     R^2 = 1 - \frac{((Y - \text{fitted\_model.fittedvalues})^2).sum()}{((Y - Y.\text{mean}())^2).sum()}
     \]
   - This expression measures the proportion of variance in \( Y \) explained by the model. It is equivalent to `fitted_model.rsquared` in statistical libraries, confirming they are the same.

3. **Interpretation of \( R^2 \) as a Measure of Model Accuracy**:
   - \( R^2 \) quantifies how well the model explains the variation in \( Y \). A higher \( R^2 \) indicates a model that explains a larger proportion of the variability in the observed data.

4. **Squared Correlation in Simple Linear Regression**:
   - In Simple Linear Regression (with one predictor), the squared correlation \( \text{correlation}(Y, x)^2 \) equals \( R^2 \).    - This relationship does not hold in multiple regression models with more than one predictor.

In summary, in Simple Linear Regression, \( R^2 \) and \( \text{correlation}(Y, x)^2 \) are equivalent because both capture the strength of the linear relationship between \( Y \) and \( x \), with \( R^2 \) representing the proportion of variation in \( Y \) explained by \( x \).

````python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
import plotly.express as px

# Set seed for reproducibility
np.random.seed(42)

# Assuming 'old_faithful' is your DataFrame and you want to filter for long wait times
long_wait_limit = 71
long_wait = old_faithful.waiting > long_wait_limit

# Fit the initial model and print the summary
initial_model = smf.ols('duration ~ waiting', data=old_faithful[long_wait]).fit()
print(initial_model.summary().tables[1])

# Step 1: Bootstrap sampling
def bootstrap_slope(data, n_iterations=1000):
    slopes = []
    for _ in range(n_iterations):
        sample = data.sample(frac=1, replace=True)
        model = smf.ols('duration ~ waiting', data=sample).fit()
        slopes.append(model.params['waiting'])
    return slopes

# Step 2: Perform bootstrap and collect slopes
bootstrap_slopes = bootstrap_slope(old_faithful[long_wait])

# Step 3: Visualize bootstrap sampling distribution using Plotly
bootstrap_fig = px.histogram(bootstrap_slopes, nbins=30, 
                              title='Bootstrap Sampling Distribution of Slope Coefficients',
                              labels={'value': 'Slope Coefficient'},
                              opacity=0.75)
bootstrap_fig.add_vline(x=0, line_color='red', line_dash='dash', 
                         annotation_text='Null Hypothesis (Slope = 0)', 
                         annotation_position='top right')
bootstrap_fig.show(renderer="png")  # Use PNG for submission

# Step 4: Simulate data under the null hypothesis
def simulate_data(n, beta_0=1.65, beta_1=0, sigma=0.37):
    x = np.random.uniform(low=2, high=5, size=n)  # Waiting times (example range)
    y = beta_0 + beta_1 * x + np.random.normal(0, sigma, size=n)  # Simulated Y
    return pd.DataFrame({'waiting': x, 'duration': y})

# Simulate samples
n = 160
simulated_slopes = []
n_simulations = 1000

for _ in range(n_simulations):
    simulated_data = simulate_data(n)
    model = smf.ols('duration ~ waiting', data=simulated_data).fit()
    simulated_slopes.append(model.params['waiting'])

# Step 5: Visualize simulated sampling distribution using Plotly
simulated_fig = px.histogram(simulated_slopes, nbins=30, 
                              title='Simulated Sampling Distribution of Slope Coefficients Under Null Hypothesis',
                              labels={'value': 'Slope Coefficient'},
                              opacity=0.75)
simulated_fig.add_vline(x=0, line_color='red', line_dash='dash', 
                         annotation_text='Null Hypothesis (Slope = 0)', 
                         annotation_position='top right')
simulated_fig.show(renderer="png")  # Use PNG for submission

# Step 6: Calculate 95% confidence interval for bootstrap slopes
conf_interval = np.percentile(bootstrap_slopes, [2.5, 97.5])
print(f"95% Bootstrap Confidence Interval: {conf_interval}")

# Step 7: Check if 0 is in the confidence interval
contains_zero = 0 >= conf_interval[0] and 0 <= conf_interval[1]
print(f"Does the 95% CI contain 0? {contains_zero}")

# Step 8: Get p-value from original model
original_p_value = initial_model.pvalues['waiting']
print(f"Original p-value from the model: {original_p_value}")

# Step 9: Calculate p-value from simulated slopes
simulated_p_value = np.mean(np.array(simulated_slopes) >= 0)
print(f"Simulated p-value: {simulated_p_value}")
````

   - The goal of bootstrapping in this context is to evaluate the variability of the slope estimates and to assess the null hypothesis by comparing the bootstrapped results to the original model's slope.
